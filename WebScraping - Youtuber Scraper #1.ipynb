```python
#impor packages yang diperlukan
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from textblob import TextBlob
from datetime import datetime, timedelta
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import time
import requests as re
import pandas as pd
```

```python
# Inisialisasi driver Chrome dan membuka URL YouTube
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get('https://www.youtube.com/@corbuzier/videos')
```

jika sudah run maka tunggu sampai chrome muncul dengan halaman youtube

lalu scroll sampai batas bawah/terakhir pada halaman youtube tersebut untuk mengambil semua data

```python
# Mengambil kode HTML dari halaman
html = driver.page_source
```

```python
# Membuat objek BeautifulSoup dari html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Mencari semua elemen video
videos = soup.find_all("div",{"id":"dismissible"})
```

```python
# Loop melalui setiap elemen video
master_list=[]
for video in videos :
    data_dict = {}
    # Mendapatkan judul video
    data_dict['title'] = video.find('yt-formatted-string',{'id':'video-title'}).text
    # Mendapatkan URL video
    data_dict['video_url'] ='https://www.youtube.com'+ video.find('a',{'id':'thumbnail'})['href']
    # Mendapatkan metadata  tentang video seperti jumlah views dan usia video
    meta = video.find('div',{'id':'metadata-line'}).find_all('span')
    data_dict['views'] = meta[0].text
    data_dict['video_age'] = meta[1].text
    
    master_list.append(data_dict)
```

```python
# Menutup browser setelah selesai scraping agar data tidak berubah
driver.quit()
```

```python
# Membuat DataFrame dari hasil web scraping
youtube_df = pd.DataFrame(master_list)
```

```python
# Mencetak DataFrame youtube_df
youtube_df
```

```python
#Membuat Fungsi untuk mengkonversi jumlah views menjadi angka numerik
def convert_views(df) :
    # Jika 'K' (ribu) ada dalam views
    if 'K' in df['views']:
        # Menghilangkan 'K' dan mengalikan dengan 1000
        views = float(df['views'].split('K')[0])*1000
        return views
    # Jika 'M' (juta) ada dalam views
    elif 'M' in df['views']:
        # Menghilangkan 'M' dan mengalikan dengan 1000000
        views = float(df['views'].split('M')[0])*1000000
        return views
```

```python
# Mengaplikasikan fungsi convert_views ke setiap baris DataFrame
youtube_df['clean_views'] = youtube_df.apply(convert_views, axis=1)
```

```python
# Mengonversi tipe data kolom clean_views menjadi integer
youtube_df['clean_views'] = youtube_df['clean_views'].astype(int)
```

```python
# Membuat sebuah fungsi yang dapat mengubah string yang berisi
# waktu relatif menjadi objek datetime
def string_to_datetime(string):
    # Mengubah string menjadi list dengan memisahkan kata-kata
    parts = string.split()

    # Mendapatkan angka dan satuan waktu
    number = int(parts[0])
    unit = parts[1]

    # Mengambil tanggal dan waktu saat ini
    now = datetime.now()

    # Mengurangi tanggal dan waktu saat ini
    #dengan angka dan satuan yang sesuai
    if unit == 'minutes':
        delta = timedelta(minutes=number)
    elif unit == 'hours':
        delta = timedelta(hours=number)
    elif unit == 'days':
        delta = timedelta(days=number)
    elif unit == 'weeks':
        delta = timedelta(weeks=number)
    elif unit == 'months':
        delta = timedelta(days=number*30)
    elif unit == 'years':
        delta = timedelta(days=number*365)
    else:
        delta = timedelta(0)

    # Menghasilkan datetime yang diinginkan
    return now - delta

# Membuat sebuah fungsi yang dapat mengubah objek datetime
# menjadi string dengan format %d-%m-%Y
def datetime_to_string(dt):
    # Menggunakan metode strftime() dengan format yang diinginkan
    return dt.strftime("%d-%m-%Y")

# Melakukan perulangan untuk setiap data di kolom video_age
for i, data in enumerate(youtube_df["video_age"]):
    # Mengubah data menjadi datetime dengan menggunakan fungsi pertama
    result = string_to_datetime(data)

    # Mengubah datetime menjadi string dengan menggunakan fungsi kedua
    result = datetime_to_string(result)

    # Menyimpan hasil ke dalam data frame
    youtube_df.loc[i, "clean_video_age"] = result
```
